# Example to plot data from a .csv file

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SW_01_TBAR6_MO.csv')
```

```python
df['z_Tbar'] += 35
df['F_Tbar'] -= 5
```

```python
x = df['F_Tbar']
y = df['z_Tbar']
```

```python
plt.scatter(x, y)
plt.xlabel('F_Tbar')
plt.ylabel('z_Tbar')
plt.xlim(0, 30)
plt.ylim(-300, 0)

# Add a grid to the plot
plt.grid(which='both', color='gray', linewidth=0.5, linestyle='--')
```

```python
df2 = pd.read_csv('SW_01_TBAR5_CY.csv')
```

```python
df2['z_Tbar'] += 35
df2['F_Tbar'] -= 5
```

```python
x2 = df2['F_Tbar']
y2 = df2['z_Tbar']
```

```python
plt.plot(x, y, label='First CSV')
plt.plot(x2, y2, color='b', linestyle='--',linewidth=0.5,label='Second CSV')
plt.xlim(0, 30)
plt.ylim(-300, 20)

# Add a grid to the plot
plt.grid(which='both', color='gray', linewidth=0.5, linestyle='--')

# Set the labels for the x and y axes
plt.xlabel('F_Tbar')
plt.ylabel('z_Tbar')
plt.legend()

# Show the plot
plt.show()
```
