```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('water.csv')
```

```python
# Extract the first two columns
y = df.iloc[:, 0]
x = df.iloc[:, 1]

# Create a scatter plot
plt.scatter(x, y)

# Adding labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of First Two Columns')

plt.ylim(350, 0)


# Display the


plt.show()
```
