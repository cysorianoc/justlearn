```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3]
y = [2, 4, 6]

# Create a scatter plot
plt.scatter(x, y)

# Adding labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of X vs Y')

# Display the plot
plt.show()
```
